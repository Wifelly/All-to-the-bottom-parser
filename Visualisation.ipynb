```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests

def bar_plot_count(category, title, logy=False):
    df = pd.DataFrame(category, columns=['count', 'country'])
    df.index = df.country
    df.plot.bar(figsize=(15, 5), logy=logy)
    plt.title(title)
    plt.show()
    
def plot_count(category, title):
    df = pd.DataFrame(category)
    df.plot(figsize=(15, 5), ylim=(90,200))
    plt.title(title)
    plt.show()
```

# Первое задание

```python
response = requests.get('http://0.0.0.0:5000/first').json()

bar_plot_count(response, 'Visits', True)
```

# Второе задание

```python
semi_manufactures = []
frozen_fish = []
canned_food = []
fresh_fish = []
caviar = []
response = requests.get('http://0.0.0.0:5000/second_visual').json()

for item in response:
    if item[1].split()[1] == 'semi_manufactures':
        semi_manufactures.append([item[0], item[1].split()[0]])
    elif item[1].split()[1] == 'frozen_fish':
        frozen_fish.append([item[0], item[1].split()[0]])
    elif item[1].split()[1] == 'canned_food':
        canned_food.append([item[0], item[1].split()[0]])
    elif item[1].split()[1] == 'fresh_fish':
        fresh_fish.append([item[0], item[1].split()[0]])
    elif item[1].split()[1] == 'caviar':
        caviar.append([item[0], item[1].split()[0]])

bar_plot_count(semi_manufactures, 'semi_manufactures')
bar_plot_count(frozen_fish, 'frozen_fish')
bar_plot_count(canned_food, 'canned_food')
bar_plot_count(fresh_fish, 'fresh_fish')
bar_plot_count(caviar, 'caviar')
```

# Третье задание

```python
response = requests.get('http://0.0.0.0:5000/third').json()
for i, item in enumerate(response):
    response[i] = [int(item.split()[2])]

canned_food = response[0:24]
caviar = response[24:48]
fresh_fish = response[48:72]
semi_manufactures = response[72:]

plot_count(canned_food, 'canned_food')
plot_count(caviar, 'caviar')
plot_count(fresh_fish, 'fresh_fish')
plot_count(semi_manufactures, 'semi_manufactures')
```
